```python
import pandas as pd

df_distritos = pd.read_csv('distritos.csv')
df_pasivos_mineros = pd.read_csv('pasivos_mineros.csv')
df_pasivos_mineros
```

```python
list_departamentos = ["AMAZONAS",
    "ANCASH",
    "APURIMAC",
    "AREQUIPA",
    "AYACUCHO",
    "CAJAMARCA",
    "CUSCO",
    "HUANCAVELICA",
    "HUANUCO",
    "ICA",
    "JUNIN",
    "LA LIBERTAD",
    "LAMBAYEQUE",
    "LIMA",
    "LORETO",
    "MADRE DE DIOS",
    "MOQUEGUA",
    "PASCO",
    "PIURA",
    "PUNO",
    "SAN MARTIN",
    "TACNA",
    "TUMBES",
    "UCAYALI"]

for department_name in list_departamentos:
    df_distritos_by_department=df_distritos[df_distritos['Departamento'].str.upper() == department_name]
    df_distritos_by_department.to_csv(f"distritos/{department_name.lower()}.csv", index=False)
    df_pasivos_mineros_by_department=df_pasivos_mineros[df_pasivos_mineros['REGIÓN'] == department_name]
    df_pasivos_mineros_by_department.to_csv(f"pasivos_mineros/{department_name.lower()}.csv", index=False)
```
